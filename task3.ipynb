```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic = sns.load_dataset('titanic')  # Or use pd.read_csv('path_to_csv')

# Display first few rows
print(titanic.head())

# Set plot style
sns.set(style='whitegrid')

# 1. Survival count
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='survived', palette='viridis')
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.show()

# 2. Survival by gender
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='sex', hue='survived', palette='pastel')
plt.title('Survival by Gender')
plt.show()

# 3. Survival by class
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='pclass', hue='survived', palette='coolwarm')
plt.title('Survival by Passenger Class')
plt.show()

# 4. Age distribution
plt.figure(figsize=(8, 5))
sns.histplot(titanic['age'].dropna(), kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()

# 5. Heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```
